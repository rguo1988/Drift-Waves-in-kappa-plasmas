# Gamma distribution $G_1$ and $G_2$

```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy import special as sp
%config InlineBackend.figure_format = 'svg'
```

```python
plt.rcParams.update({'font.size': 16})
plt.rcParams.update({'lines.linewidth': 2})
```

```python
b = np.linspace(0,5,num=200)
kappa = 3
g1 = kappa/sp.gamma(kappa+0.5) * (kappa*b)**(kappa-0.5) * np.exp(-kappa*b)
g2 = kappa*(kappa-1.5)/sp.gamma(kappa+0.5) * (kappa*b)**(kappa-1.5) * np.exp(-kappa*b)
```

```python
fig = plt.figure(figsize=(6,4),layout='constrained')
plt.plot(b,g1,linestyle='-',color='tab:blue',label=r'$G_1$')
plt.plot(b,g2,linestyle='--',color='tab:red',label=r'$G_2$')
plt.xlabel(r'$b_\sigma$')
plt.grid()
plt.legend()
plt.savefig('./fig4_g1g2_kappa=3.pdf')
plt.show()
```
